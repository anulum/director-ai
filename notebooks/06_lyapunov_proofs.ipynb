# 06 — Lyapunov Stability Proofs

Runs the symbolic (SymPy) and numerical (Monte Carlo) stability
proofs for the SEC Lyapunov functional.

```python
from director_ai.research.physics.lyapunov_proof import run_all_proofs, ProofResult
```

## Run All Proofs

```python
proofs = run_all_proofs()

print(f"{'Proof':<25} {'Verified':>8}  Statement")
print('─' * 80)
for p in proofs:
    status = '✓ PASS' if p.verified else '✗ FAIL'
    print(f"{p.name:<25} {status:>8}  {p.statement}")
```

## Proof Details

```python
for p in proofs:
    print(f"\n{'='*60}")
    print(f"Proof: {p.name}")
    print(f"Statement: {p.statement}")
    print(f"Verified: {p.verified}")
    if p.symbolic_expr:
        print(f"Symbolic: {p.symbolic_expr}")
    if p.detail:
        print(f"Detail: {p.detail}")
```

## Summary

```python
n_pass = sum(1 for p in proofs if p.verified)
n_total = len(proofs)
print(f"\n{n_pass}/{n_total} proofs verified.")
if n_pass == n_total:
    print("All Lyapunov stability conditions satisfied. QED.")
```
